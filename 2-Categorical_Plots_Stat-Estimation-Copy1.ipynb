### Countplot()

A simple plot, it merely shows the total count of rows per category. 

```python
sns.countplot(x='division',data=df)
```

```python
sns.countplot(x='level of education',data=df)
```

### Breakdown within another category with 'hue'

```python
sns.countplot(x='level of education',data=df,hue='training level')
```

### barplot()

A bar plot shows catergorical data as rectangular bars with the height of bars proportional to the value they represent. It is often used to compare between values of different categories in the data.

```python
# By default barplot() will show the mean
sns.barplot(x='level of education',y='salary',data=df,estimator=np.mean,ci='sd')
```

```python
sns.barplot(x='level of education',y='salary',data=df,estimator=np.mean,ci='sd',hue='division')
plt.legend(bbox_to_anchor=(1, 1.05))
```

## Categorical Plots - Distribution of observation within Categories

We already know about distplot() which allows to view the distribution of a single feature, now we will break down that same distribution per category.

```python
df1 = pd.read_csv("StudentsPerformance.csv")
```

```python
df1.head()
```

### Boxplot
boxplot display distribution through the use of quartiles and an IQR for outliers.

```python
sns.boxplot(x='parental level of education',y='math score',data=df1)
```

from above we can see that there are some outliers present in "bachelor's degree", "some college", "high school" and "some high school".

```python
# Adding hue for further segmentation

sns.boxplot(x='parental level of education',y='math score',data=df1,hue='gender')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
```

```python
# WE Can SWITCH X AND Y FOR THE ORIENTATION TO MAKE SENSE!
sns.boxplot(x='math score',y='parental level of education',data=df1,orient='h')
```

```python
# we can change the width of the boxplot

sns.boxplot(x='parental level of education',y='math score',data=df1,hue='gender',width=0.3)
```

### Violinplot

A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

```python
sns.violinplot(x='parental level of education',y='math score',data=df1, palette='coolwarm')
```

```python
sns.violinplot(x='parental level of education',y='math score',data=df1,hue='gender', palette='viridis')
```

```python
sns.violinplot(x='parental level of education',y='writing score',data=df1,hue='gender', palette='Set3')
```

**split**

When using hue nesting with a variable that takes two levels, setting split to True will draw half of a violin for each level. This can make it easier to directly compare the distributions.

```python
sns.violinplot(x='parental level of education',y='math score',data=df1,hue='gender',split=True, palette='Set3')
```

**inner**
 
Representation of the datapoints in the violin interior. If box, we can draw a miniature boxplot. If quartiles, we can draw the quartiles of the distribution. If point or stick, we can show each underlying datapoint. Using None will draw unadorned violins.

```python
plt.style.use('ggplot')
```

```python
sns.violinplot(x='parental level of education',y='math score',data=df1,inner=None)
```

```python
sns.violinplot(x='parental level of education',y='math score',data=df1,inner='box')
```

```python
sns.violinplot(x='parental level of education',y='math score',data=df1,inner='quartile')
```

```python
sns.violinplot(x='parental level of education',y='math score',data=df1,inner='stick')
```

```python
sns.violinplot(x='parental level of education',y='reading score',data=df1,inner='stick')
```

**orientation**

we can simply switch the continuous variable to y and the categorical to x


```python
sns.violinplot(x='math score',y='parental level of education',data=df1)
```

```python
sns.violinplot(x='writing score',y='parental level of education',data=df1)
```

```python
#adding hue

sns.violinplot(x='writing score',y='parental level of education',hue='gender',data=df1,split=True)
```

**bandwidth**

It is Similar to bandwidth argument for kdeplot

```python
sns.violinplot(x='parental level of education',y='math score',data=df1,bw=0.1)
```

We can use a some advance plot like **boxenplot** and **swarmplot** to achieve the same effect as the boxplot and violinplot, but with slightly more information included.

```python
sns.swarmplot(x='reading score',data=df1)
```

```python
sns.swarmplot(x='math score',y='race/ethnicity',data=df1)
```

```python
sns.swarmplot(x='race/ethnicity',y='math score',data=df1)
```

```python
sns.swarmplot(x='race/ethnicity',y='math score',data=df1,hue='gender')
```

```python
sns.swarmplot(x='race/ethnicity',y='math score',data=df1,hue='gender',dodge=True)
```

### boxenplot (letter-value plot)

This style of plot was originally named a “letter value” plot because it shows a large number of quantiles that are defined as “letter values”. It is similar to a box plot in plotting a nonparametric representation of a distribution in which all features correspond to actual observations. By plotting more quantiles, it provides more information about the shape of the distribution, particularly in the tails.

```python
sns.boxenplot(x='math score',y='race/ethnicity',data=df1)
```

```python
sns.boxenplot(x='race/ethnicity',y='math score',data=df1)
```

```python
sns.boxenplot(x='race/ethnicity',y='writing score',data=df1,hue='gender')
```
